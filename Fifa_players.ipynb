```python
columns_to_drop = [
    'ID',                    # Player ID is usually not necessary for analysis.
    'date_id',               # Date ID typically does not provide useful information for analysis.
    'Best overall',          # This column often overlaps with other metrics and does not add significant value.
    'Growth',                # Growth information is generally not used in performance analysis.
    'Wage',                  # Salary is typically a financial metric that doesn't contribute to performance analysis.
    'Release clause',        # Release clause is usually irrelevant for player performance evaluation.
    'Weak foot',             # Weak foot ability often overlaps with other attributes.
    'Skill moves',# Skill moves usually do not provide critical insights for performance analysis.
    'International reputation',
    'Attacking  work rate',
    'Defensive work rate',
    'Body type',                       
    'Number of',  
    'player_name', 'Curve', 'CurveTotal skill', 'Short passing',
    'Potential', 'GK Positioning', 'GK Reflexes', 'Weak foot', 'Total goalkeeping','Shot power', 'Agility', 'Long passing',
    'International reputation','Value','Growth', 'Wage', 'Release clause', 'Volleys','Sprint speed', 'Shot power','Strength','Long shots', 'Vision', 'Composure', 'Standing tackle', 'Sliding tackle', 'GK Diving', 'GK Kicking', 'Skill moves'
    
]
```

```python
df_new = df.drop(columns=columns_to_drop)
df_new.head()
```

```python
df_new.info()
```

```python
df_new.replace({'Yes': 1, 'No':0}, inplace=True)
#for right food:1, for left foot:0
df_new.replace({'Left': 0, 'Right': 1}, inplace=True)
```

```python
df_new.head()
```

```python
for col in df_new.columns:
    if df_new[col].dtype == 'object': 
        df_new[col] = pd.to_numeric(df_new[col], errors='coerce')
```

```python
numeric_df = df_new.select_dtypes(include=['number'])
numeric_df.corr()
```

```python
df_new['foot'] = df_new['foot'].astype(int)
df_new = df_new[df_new['foot'].isin([0, 1])]
```

```python
sns.countplot(data=df_new, x='foot', hue='foot')
plt.xticks([0, 1], ['Left Foot', 'Right Foot']) 
plt.title('sdasda')
plt.show()
```
